```python
# Import necessary libraries
import numpy as np
import pymilvus

# Connect to Milvus server
from pymilvus import connections, Milvus
from pymilvus.exceptions import MilvusException
import grpc

def connect_to_milvus_server(host='localhost', port=19530, timeout=30):
    try:
        connections.connect(host=host, port=port, timeout=timeout)
        milvus = Milvus()
        print("Connected to Milvus server successfully!")
        return milvus
    except grpc.FutureTimeoutError as e:
        raise MilvusException(
            code=2,
            message=f"Failed connecting to server on {host}:{port}. Timeout",
        )
    except MilvusException as e:
        print(f"MilvusException: {e}")

if __name__ == "__main__":
    try:
        milvus_client = connect_to_milvus_server()
        if milvus_client is not None:
            # Define Milvus collection
            collection_name = "embedding_collection"
            schema = pymilvus.CollectionSchema(
                fields=[
                    pymilvus.FieldSchema(name="idNumber", dtype=pymilvus.DataType.INT64, is_primary=True),  # Assuming 'idNumber' is your primary key field
                    pymilvus.FieldSchema(name="embedding", dtype=pymilvus.DataType.FLOAT_VECTOR, dim=np.array(combined_vector).shape[0])
                ]
            )

            # Create collection if not exists
            if not milvus_client.has_collection(collection_name):
                milvus_client.create_collection(collection_name, schema)

            # Insert data into Milvus collection
            for id_number, embedding in embeddings.items():
                milvus_client.insert(collection_name, [embedding], ids=[id_number])

            print("Data inserted into Milvus collection successfully!")
        else:
            print("Failed to connect to Milvus server.")

    except Exception as e:
        print(f"Error: {e}")
    finally:
        if milvus_client is not None:
            # Disconnect from Milvus server
            milvus_client.disconnect()
```
